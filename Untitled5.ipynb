```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Создание модели
def create_model():
    model = keras.Sequential([
        layers.Dense(32, activation='relu', input_shape=(X_scaled.shape[1],)),
        layers.Dense(16, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# 4. Кросс-валидация
kf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
fold_no = 1

for train_index, test_index in kf.split(X_scaled, y):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model = create_model()

    # Обучение модели
    history = model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test), verbose=0)

    # 5. Построение графиков
    plt.figure(figsize=(12, 5))

    # График точности
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Точность (обучающая)')
    plt.plot(history.history['val_accuracy'], label='Точность (валидационная)')
    plt.title(f'Фолд {fold_no}: Точность модели')
    plt.xlabel('Эпохи')
    plt.ylabel('Точность')
    plt.legend()

    # График потерь
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Потери (обучающая)')
    plt.plot(history.history['val_loss'], label='Потери (валидационная)')
    plt.title(f'Фолд {fold_no}: Потери модели')
    plt.xlabel('Эпохи')
    plt.ylabel('Потери')
    plt.legend()

    plt.tight_layout()
    plt.show()

    # 6. Отчет о классификации и матрица путаницы
    y_pred = (model.predict(X_test) > 0.5).astype("int32")
    print(f'Accuracy: {accuracy_score(y_test, y_pred):.2f}')
    print(f'Фолд {fold_no}:n')
    print(classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred)
    print(f'Матрица путаницы для фолда {fold_no}:n{cm}n')

    # Расшифровка матрицы ошибок
    tn, fp, fn, tp = cm.ravel()

    print(f'Модель предсказала, что выживет, а по факту не выжило: {fp}')
    print(f'Модель предсказала, что не выживет, а по факту выжило: {fn}')
    print(f'Модель предсказала, что выживет, и по факту выжило: {tp}')
    print(f'Модель предсказала, что не выживет, и по факту не выжило: {tn}n')

    fold_no += 1
```

```python
kf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
accuracy_history = []
loss_history = []

for train_index, test_index in kf.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Создаем модель
    model = keras.Sequential([
        layers.Dense(10, activation='relu', input_shape=(X_train.shape[1],)),
        layers.Dense(8, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Обучаем модель
    history = model.fit(X_train, y_train, epochs=50, validation_split=0.2, verbose=0)

    # Сохраняем историю
    accuracy_history.append(history.history['accuracy'])
    loss_history.append(history.history['loss'])

    # Прогнозирование и оценка
    y_pred = (model.predict(X_test) > 0.5).astype("int32")
    print(classification_report(y_test, y_pred))
    cm = confusion_matrix(y_test, y_pred)
    print(f'Accuracy: {accuracy_score(y_test, y_pred):.2f}')
    print("Confusion Matrix:\n", cm)

    tn, fp, fn, tp = cm.ravel()
    print(f"Модель предсказала, что погибнет, а по факту не погиб: {fp}")
    print(f"Модель предсказала, что выживет, а по факту погиб: {fn}")
    print(f"Модель предсказала, что выживет, и по факту выжила: {tp}")
    print(f"Модель предсказала, что погибнет, и по факту погиб: {tn}")

# Подготовка графиков
plt.figure(figsize=(14, 5))

# Графики для accuracy
plt.subplot(1, 2, 1)
for history_accuracy in accuracy_history:
    plt.plot(history_accuracy)
plt.title('Accuracy на обучающем наборе')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')

# Графики для loss
plt.subplot(1, 2, 2)
for history_loss in loss_history:
    plt.plot(history_loss)
plt.title('Loss на обучающем наборе')
plt.ylabel('Loss')
plt.xlabel('Epochs')

plt.tight_layout()
plt.show()
```
